```python
df["Date"] = pd.to_datetime(
    df["Date"],
    format="%Y-%m-%d",
    errors="raise"
)
```

```python
df["Date"].dtype
df["Date"].isna().sum()
```

```python
df["Year"]=df["Date"].dt.year
df["Year"].head(15)
```

```python
df["Month"]=df["Date"].dt.month
df["Month"].head(15)
```

```python
df["Weekoftheyear"]=df["Date"].dt.isocalendar().week.astype(int)
df["Weekoftheyear"].head(15)
```

```python
df["Macro_Trend"] = (df["CPI"].rank(pct=True) + df["Unemployment"].rank(pct=True))
df["Macro_Trend"].head(15)
```

```python
df = df.sort_values(["Store", "Dept", "Date"])
# WeekBeforeHoliday = 1 means this week occurs immediately before a holiday week.
df["WeekBeforeHoliday"] = (
    df.groupby(["Store", "Dept"])["IsHoliday_x"]
      .shift(-1)
      .fillna(0)
      .astype(int)
)
# WeekAfterHoliday = 1 means this week occurs immediately after a holiday week.
df["WeekAfterHoliday"] = (
    df.groupby(["Store", "Dept"])["IsHoliday_x"]
      .shift(1)
      .fillna(0)
      .astype(int)
)
```

```python
df["WeekBeforeHoliday"].head(15)
```

```python
df["WeekAfterHoliday"].head(15)
```

```python
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
for col in markdown_cols:
    df[f"{col}_missing"] = df[col].isna().astype(int)
df[markdown_cols] = df[markdown_cols].fillna(0)
df["Total_MarkDown"] = df[markdown_cols].sum(axis=1)
df["Any_MarkDown"] = (df["Total_MarkDown"] > 0).astype(int)
```

```python
df.to_csv("test_features.csv",index=False)
```
